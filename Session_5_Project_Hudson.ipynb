Download another Pokemon dataset from here :https://www.dropbox.com/s/ms0tixxeflq1toc/Pokemon.csv?dl=0 
Lets explore:

```python
import pandas as pd
df = pd.read_csv('Pokemon.csv')
df.head()
```

```python
#Let's rename the # column to id, and convert all column labels to lower case.
# use df.rename to rename the # column to id
# use df.columns.str.lower() to convert all column labels to lower case
# you should see:
df.rename(index=str, columns={"#": "id"}, inplace='TRUE')
df.columns = df.columns.str.lower()
df.head()
```

```python
#Now let's take a look at any duplicated rows via the id column.

# use df.duplicated('id', keep=False) filter
df[df.duplicated('id', keep=False)]
```

```python
# Lets exclude these duplicates in our analysis except for the first instance.

# use df.drop_duplicates
# you should see no duplictea anymore if you check :
df.drop_duplicates(subset ='id', inplace=True)
df.head()
```

```python
import numpy as np
df['type 2']
```

```python
# Fill the Nan values
df.fillna("No-type", inplace=True)
```

```python
#The dataset contains both information regarding the identity and statistics of each Pokemon species; therefore, 
#let's separate these two observational units into separate tables: pokedex and statistics.
# create pokedex with these columns: ['id', 'name', 'type 1', 'type 2', 'generation', 'legendary']
pokedex_att = ['id', 'name', 'type 1', 'type 2', 'generation', 'legendary']
pokedex = df[pokedex_att].copy()
pokedex.head()
```

```python
# create poke_statistics with these columns:['id', 'hp', 'attack', 'defense', 'sp. atk', 'sp. def', 'speed','total']
poke_stats_att = ['id', 'hp', 'attack', 'defense', 'sp. atk', 'sp. def', 'speed','total']
poke_statistics = df[poke_stats_att].copy()
poke_statistics.head()
```

```python
# Create a new df called "dual_types" by getting rid of the records for which we do not have Type 2
# You should see this :

#pokedex.head(pokedex['type 2'] != "No-type")
dual_types = pokedex[pokedex['type 2'] != "No-type"].copy()
dual_types.head()
```

```python
# Get a grid of values from 'type 1' 'type 2'  (Hint: use groupby, size, unstack, values 
#and any other attributes you may need !)
dual_type_grid = dual_types.groupby([dual_types['type 1'], dual_types['type 2']])

#dual_type_grid.id.count().plot(kind='bar')
#dual_type_grid[dual_type_grid['type 1'=="Bug"]].describe()

all_type1
for t1 in all_type1:
    dual_types_fixedtype1 = dual_types[dual_types['type 1'] == t1]
    print(dual_types_fixedtype1.groupby(dual_types_fixedtype1['type 2']).id.count().plot(kind='bar'))
    plt.title("type 1 = "+t1)
    plt.show()
#plt.show()

#Histograms of type 2 for every kind of type_1
```

```python
#Using matplotlib for heatmap

grid = dual_type_grid.count()['id'].unstack().fillna(0)
#type 1 is row
#type 2 is column
y_ticks = grid.index.values
x_ticks = grid.columns.values

print(y_ticks)
print(x_ticks)

sizey = len(y_ticks)
sizex = len(x_ticks)

aux_vecy = np.arange(0,sizey)
aux_vecx = np.arange(0,sizex)


plt.imshow(grid, cmap = 'Reds', )
plt.colorbar()
plt.xticks(aux_vecx, x_ticks, rotation=90)
plt.yticks(aux_vecy, y_ticks)
plt.xlabel("type 2")
plt.ylabel("type 1")
plt.show()
```
